```python
import pandas as pd
```

```python
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

Discover how many tables there are on the wiki page:

```python
len(df)
```

Find the table out of the 3 using indexing the cooresonds to the desired wiki table:

```python
df[0] 
```

```python
Toronto = df[0]
```

Select the rows within the Toronto table that contain 'Not assigned' for the column 'Borough'

```python
indexNames = Toronto[ Toronto['Borough'] == 'Not assigned' ].index
```

Delete these selected rows from the table:

```python
Toronto.drop(indexNames , inplace=True)
```

Show the new table making sure there are no rows with the values 'Not assigned':

```python
Toronto
```

Show how many rows are in the able:

```python
Toronto.shape
```

Sort all values by postal code column to make sure they are in asending order so they stack up with geodata CSV

```python
Toronto2 = Toronto.sort_values(by='Postal Code', axis=0, ascending=True)
Toronto2
```

Check to make sure postal code column is in ascending order

```python
Toronto_postal_list = Toronto2['Postal Code'].tolist()
Toronto_postal_list
```

Read in CSV file of geodata

```python
geodata = pd.read_csv('https://cocl.us/Geospatial_data')
geodata.head()
```

Select rows in geodata CSV that coorespond to the toronto postal codes array

```python
Tor_lat_lng = geodata.loc[geodata['Postal Code'].isin(Toronto_postal_list)]
Tor_lat_lng
```

Convert column values of tables: toronto2 & Tor_lat_lng into list arrays

```python
col1 = Toronto2['Postal Code'].to_list()
col2 = Toronto2['Borough'].to_list()
col3 = Toronto2['Neighbourhood'].to_list()
col4 = Tor_lat_lng['Latitude'].to_list()
col5 = Tor_lat_lng['Longitude'].to_list()  
```

Create final table by combinging all list arrays as columns

```python
final_table = pd.DataFrame(list(zip(col1, col2,col3, col4, col5)), 
               columns =['Postal Code', 'Borough', 'Neighbourhood', 'Latitude', 'Longitude']) 
final_table 
```
